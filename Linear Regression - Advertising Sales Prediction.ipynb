```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Import Dataset

```python
dataset=pd.read_csv("C:/Users/shaik/advertising.csv")
dataset
```

```python
dataset.shape
```

# Slicing data set - Removing the columns

```python
dataset=dataset.drop(['Radio','Newspaper'],axis=1)
dataset
```

```python
y=dataset.iloc[:,-1].values.reshape(-1,1)
y.shape
```

```python
y
```

```python
x=dataset.iloc[:,0].values.reshape(-1,1)
x
```

```python
plt.scatter(x,y)
plt.show()
```

# Split the dataset into Training and Testing Set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.30,random_state=123)
```

# Checking Training and Testing Dataset

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Import the function for Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

# Prediction of sales

```python
y_pred = lm.predict(x_test)
```

```python
y_pred
```

# Check the Prediction

```python
check = pd.DataFrame(x_test,columns=['Tv'])
check
```

```python
check['Sales_actual']=y_test
check
```

```python
check['Sales_Predicted']=y_pred
check
```

```python
check.to_csv('error_csv')
```

# Check Using a linear Regression Line

```python
plt.scatter(x,y,color='blue')
plt.plot(x_test,y_pred,color='red')
```

```python
from sklearn.metrics import mean_squared_error
import math
rmse = math.sqrt(mean_squared_error(y_test,y_pred))
rmse
```

# Accuracy

```python
lm.score(x_train,y_train)
```
